# WORD2VEC resolved

```python
from gensim import models

w2vmodel = models.KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin.gz', binary=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_train.data, data_train.target, test_size=0.2, random_state=42)
```

```python
# define function to convert documents to word2vec vectors
def embedding_feats(list_of_lists):
    DIMENSION = 300
    zero_vector = np.zeros(DIMENSION)
    feats = []
    for tokens in list_of_lists:
        feats_for_this = np.zeros(DIMENSION)
        count_for_this = 0
        for token in tokens:
            if token in w2vmodel:
                feats_for_this += w2vmodel[token]
                count_for_this += 1
        feats.append(feats_for_this/count_for_this)
    return feats
```

```python
X_train_feats = embedding_feats([doc.split() for doc in X_train])
X_test_feats = embedding_feats([doc.split() for doc in X_test])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X_train_feats_scaled = scaler.fit_transform(X_train_feats)
X_test_feats_scaled = scaler.fit_transform(X_test_feats)
```

```python
nb_w2v = MultinomialNB()
nb_w2v.fit(X_train_feats_scaled, y_train)
```

```python
lr_w2v = LogisticRegression()
lr_w2v.fit(X_train_feats_scaled, y_train)
```

```python
svc_w2v = SVC()
svc_w2v.fit(X_train_feats_scaled,y_train)
```

```python
dt_cv = DecisionTreeClassifier()
dt_cv.fit(X_train_feats_scaled, y_train)
```

```python
y_pred = nb_w2v.predict(X_test_feats_scaled)
print("Multinomial Naive Bayes with Word2Vec Test Accuracy:", accuracy_score(y_test, y_pred))
y_pred = lr_w2v.predict(X_test_feats_scaled)
print("Logistic Regression with Word2Vec Test Accuracy:", accuracy_score(y_test, y_pred))
svc_pred = svc_w2v.predict(X_test_feats_scaled)
print("Support vector Machine with Word2Vec Test Accuracy:", accuracy_score(y_test, svc_pred))
y_pred = dt_cv.predict(X_test_feats_scaled)
print("Decision Trees with Word2Vec Test Accuracy:", accuracy_score(y_test, y_pred))
```
